```python
 # Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)
purchase_data_df.head()
```

```python
total_playrs = [{"Total Players" : purchase_data_df["SN"].nunique()}]

total_playrs_df = pd.DataFrame(total_playrs)
total_playrs_df
```

```python
unique_items =purchase_data_df["Item ID"].nunique()
avg_prch_price =round(purchase_data_df["Price"].mean(),2)
tot_prch_items =purchase_data_df["Purchase ID"].count()
tot_revenue=round(purchase_data_df["Price"].sum(),2)
#pd.options.display.float_format = '${:,.2f}'.format
summary = [{"Unique Item Count" : unique_items, "Avg Purchase Price": avg_prch_price , "Total Purchase Volume":tot_prch_items, "Total Revenue" :tot_revenue }]
summary_df = pd.DataFrame(summary)
summary_df
```

```python
purchase_data1= purchase_data_df
total_players =  purchase_data1["SN"].nunique()
#usa_ufo_df = clean_ufo_df.loc[clean_ufo_df["country"] == "us", :]
purchase_data_mod = purchase_data1.loc[purchase_data1["Gender"] == "Male", :]
purchase_data_mod2 = purchase_data1.loc[purchase_data1["Gender"] == "Female", :]

purchase_data_mod3 = purchase_data1.loc[purchase_data1["Gender"] == "Other / Non-Disclosed", :]
male_players = purchase_data_mod["SN"].nunique()
female_players = purchase_data_mod2["SN"].nunique()
other_ndisc_players = purchase_data_mod3["SN"].nunique()

percent_male = round((male_players/total_players)*100,2)
percent_female=round((female_players/total_players)*100,2)
percent_other=round((other_ndisc_players/total_players)*100,2)

summary_df =pd.DataFrame( { "" : ["Male", "Female", "Others"],"Total Count" : [male_players, female_players,other_ndisc_players],
             "Percent Value" : [percent_male,percent_female,percent_other]})



summary_df 
```

```python


grouped_purchase_data_df = purchase_data1.groupby(['Gender'])
purchase_count = grouped_purchase_data_df["Purchase ID"].count()
avg_purchase_price =grouped_purchase_data_df["Price"].mean()
total_purchase_value =grouped_purchase_data_df["Price"].sum()
#avg_total_purchase_per_person = grouped_purchase_data_df.groupby("SN")["Price"].mean()

normalized_total = total_purchase_value / [female_players, male_players,other_ndisc_players]

summary_df = pd.DataFrame({"Purchase Count" : purchase_count, "Average Purchase Price" : avg_purchase_price ,"Total Purchase Value" :total_purchase_value, "Average Total Purchase Per Person" : normalized_total })

summary_df 
```

```python
purchase_data3= purchase_data_df

bins=[0,9,14,19,24,29,34,39,100]
grp_name=["<=10", "11 to 15", "16 to 20", "21 to 25", "26 to 30", "31 to 35",
                "35 to 40", "> 40"]
purchase_data3 ["Total Counts"] =pd.cut(purchase_data3["Age"],bins,labels=grp_name)



ranges = purchase_data3 ["Total Counts"].value_counts()

percents = round((purchase_data3 ["Total Counts"].value_counts()/purchase_data3["SN"].nunique())*100,2)

summary_df = pd.DataFrame({ "Total Count" : ranges  ,"Percent Of Players" :percents })

summary_df
```

```python
del purchase_data3 ["Total Counts"]
purchase_data3.head()
```

```python
purchase_data4 = purchase_data3.groupby(["Total Count"])

purc_count = purchase_data4["Purchase ID"].count()
avg_purc_price = round( purchase_data4 ["Price"].mean() ,2)
total_purch_value = purchase_data4 ["Price"].sum()

avg_total_purch_per_person = round( purchase_data4 ["Price"].sum() / purchase_data4["SN"].nunique(),2)


summary_df = pd.DataFrame({ "Purchase Count" : purc_count  ,"Average Purchase Price" :avg_purc_price,"Total Purchase Value" : total_purch_value , "Avg Total Purchase per person" :avg_total_purch_per_person  })

summary_df.head(10)
```
